```python
# Find cloest k defenders around a shot
def find_dist(shot_location, frame, k):
    dist_list = []
    if not isinstance(frame, list):
        return None
    for player in frame:
        if player['teammate'] == False:
            dist = round(((shot_location[0] - player['location'][0])**2 + (shot_location[1] - player['location'][1])**2)**0.5, 2)
            dist_list.append(dist)
    dist_list.sort()
    return dist_list[:k]
```

```python
# compute closest defenders distance
df1['distances'] = df1.apply(lambda row:find_dist(row['location'], row['shot_freeze_frame'], 3), axis=1)
```

```python
def find_goal_dist(shot_location):
    return round(((120 - shot_location[0])**2 + (40 - shot_location[1])**2)**0.5, 2)
```

```python
# compute shot from goal distance
df1['goal_dist'] = df1.location.apply(lambda x:find_goal_dist(x))
```

```python
import numpy as np

def find_angle_to_goal(shot_location):
    # Define the vectors from shot location to each goalpost
    vector_left = (120 - shot_location[0], 36 - shot_location[1])
    vector_right = (120 - shot_location[0], 44 - shot_location[1])
    
    # Calculate the dot product of the two vectors
    dot_product = vector_left[0] * vector_right[0] + vector_left[1] * vector_right[1]
    
    # Calculate the magnitudes of the two vectors
    magnitude_left = np.sqrt(vector_left[0]**2 + vector_left[1]**2)
    magnitude_right = np.sqrt(vector_right[0]**2 + vector_right[1]**2)
    
    # Calculate the cosine of the angle
    cos_theta = dot_product / (magnitude_left * magnitude_right)
    
    # Compute the angle in radians and then convert to degrees
    angle_radians = np.arccos(np.clip(cos_theta, -1.0, 1.0))  # Clip for numerical stability
    angle_degrees = np.degrees(angle_radians)
    
    return round(angle_degrees, 2)
```

```python
# compute angle to goal
df1['angle'] = df1.location.apply(lambda x:find_angle_to_goal(x))
```

## Build the first PYMC model

```python
df_try1 = df1.copy()
df_try1["Intercept"] = np.ones(len(df_try1))

from sklearn.model_selection import train_test_split

# X = df_try1[['Intercept', 'shot_aerial_won', 'shot_first_time', 'under_pressure', 'shot_one_on_one', 'shot_open_goal', 'def_dist', 'goal_dist', 'angle']]
X = df_try1[['Intercept', 'under_pressure', 'shot_one_on_one', 'shot_open_goal', 'def_dist', 'goal_dist', 'angle']]
y = df_try1[['shot_outcome']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5, stratify=y)

from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# Define columns for binary encoding, one-hot encoding, and scaling
# binary_columns = ['shot_aerial_won', 'shot_first_time', 'under_pressure', 'shot_one_on_one', 'shot_open_goal']
binary_columns = ['under_pressure', 'shot_one_on_one', 'shot_open_goal']
numeric_columns = ['def_dist', 'goal_dist', 'angle']

# Set up the transformations for each type of column
binary_encoder = Pipeline(steps=[
    ('one_hot_binary', OneHotEncoder(drop='if_binary', sparse_output=False))
])

scaler = Pipeline(steps=[
    ('scale', StandardScaler())
])

# Apply the transformations only on X_train
preprocessor = ColumnTransformer(
    transformers=[
        ('binary', binary_encoder, binary_columns),
        ('scaler', scaler, numeric_columns)
    ],
    remainder='passthrough'
    # remainder='drop'
)

# Fit and transform X_train, then transform X_test
X_train_processed_1 = preprocessor.fit_transform(X_train)
X_test_processed_1 = preprocessor.transform(X_test)
```

```python
# labels = ["Intercept", 'shot_aerial_won', 'shot_first_time', 'under_pressure', 
        #   'shot_one_on_one', 'shot_open_goal', "def_dist", "goal_dist", "angle"]
labels = ["Intercept", 'under_pressure', 'shot_one_on_one', 'shot_open_goal', "def_dist", "goal_dist", "angle"]

coords = {"coeffs": labels}

with pm.Model(coords=coords) as model:
    # data containers
    X = pm.MutableData("X", X_train_processed_1)
    y = pm.MutableData("y", y_train.shot_outcome)
    # priors
    b = pm.Normal("b", mu=0, sigma=1, dims="coeffs")
    # linear model
    mu = pm.math.dot(X, b)
    # link function
    p = pm.Deterministic("p", pm.math.sigmoid(mu))
    # likelihood
    pm.Bernoulli("obs", p=p, observed=y)

pm.model_to_graphviz(model)
```
